## Введение

В качестве исходных данных были взяты результаты экспериментальных исследований, проводимых в целях улучшения технологического процесса электронно-лучевой сварки изделия, сборка которого состоит из элементов, состоящих из разнородного материала. Установка электронно-лучевой сварки, на которой проводились исследования, предназначена для сварки электронным лучом в глубоком вакууме деталей сборочных единиц из нержавеющих сталей, титановых, алюминиевых и специальных сплавов. Существующая установка электронно-лучевой сварки обеспечивает повторяемость режимов в рамках возможностей реализованной системы управления. Работы по сварке выполнялись на образцах-имитаторах, соответствующих технологическому изделию. Для уменьшения вложения энергии при сварке:
1.	Снижалась величина сварочного тока (IW);
2.	Увеличивался ток фокусировки электронного пучка (IF);
3.	Увеличивалась скорость сварки (VW);
4.	Менялось расстояние от поверхности образцов до электронно-оптической системы (FP). 


По совокупности параметров технологических режимов обеспечивались минимально возможные размеры сварных швов: глубина шва (Depth) и ширина шва (Width).

В процессе выполнения работ была произведена электронно-лучевая сварка 18-ти единиц образцов. Результаты металлографического контроля по размерам сварного шва для каждого образца проводились в 4-х поперечных сечениях сварного шва. Ускоряющее напряжение было постоянным в диапазоне 19,8 – 20 кВ. Набор полученных данных собраны в составе режимов сварки, размеров сварочных швов в поперечных сечениях всех образцов.

Требуется: провести прогнозирование глубины (Depth) и ширины (Width) сварного шва в зависимости от параметров технологического процесса (IW, IF, VW, FP).

## Моделирование

Мы имеем 2 целевых переменных, 4 признака и 65 наблюдений, на основе которых будем создавать прогнозную модель.

Отметим, что количество наблюдений кажется недостаточным для построения достаточно точной модели. Но в случае, когда процесс получения экспериментальных данных является дорогим и/или ресурсоемким, приходится оперировать теми данными, которые удалось добыть. С другой стороны, однозначного ответа на вопрос, что считать достаточно большой выборкой, нет. Например, есть эмпирическое правило, которое гласит, что минимально необходимое количество наблюдений в выборке линейно связано с числом признаков, и минимальная выборка может быть рассчитана по формуле k * n, где n — число признаков для каждого наблюдения, а k — коэффициент, который из опыта часто считают равным 10. С этой точки зрения нам достаточно имеющихся данных. 

В то же время разные семейства алгоритмов могут быть более или менее эффективны для разного размера выборок. Например, нейронные сети могут хуже работают на малых данных — для их обучения часто используют сотни тысяч наблюдений.

В данном случае нам предстоит решить задачу обучения с учителем, а точнее — задачу регрессии. Для ее решения мы попробуем применить несколько разных методов, а именно:
- Алгоритм случайного леса (RandomForestRegressor)
- Линейная регрессия (Lasso и Ridge)
- Градиентный бустинг (GradientBoostingRegressor)
- Метод опорных векторов (SVR)
- Метод k-ближайших соседей (KNeighborsRegressor)
- Нейронная сеть

Качество моделей будем оценивать с помощью коэффициента детерминации R<sup>2</sup>. Эта метрика принимает значения от 0 до 1 и показывает, насколько сделанный моделью прогноз отражает целевую переменную. Будем искать модель с наибольшим R<sup>2</sup>, и вместе с тем оценим еще две важные метрики: среднюю абсолютную ошибку MAE и среднюю квадратическую ошибку MSE.

```python
# класс для попарной отрисовки графиков
class SeabornFig2Grid():

    def __init__(self, seaborngrid, fig,  subplot_spec):
        self.fig = fig
        self.sg = seaborngrid
        self.subplot = subplot_spec
        if isinstance(self.sg, sns.axisgrid.FacetGrid) or \
            isinstance(self.sg, sns.axisgrid.PairGrid):
            self._movegrid()
        elif isinstance(self.sg, sns.axisgrid.JointGrid):
            self._movejointgrid()
        self._finalize()

    def _movegrid(self):
        """ Move PairGrid or Facetgrid """
        self._resize()
        n = self.sg.axes.shape[0]
        m = self.sg.axes.shape[1]
        self.subgrid = gridspec.GridSpecFromSubplotSpec(n,m, subplot_spec=self.subplot)
        for i in range(n):
            for j in range(m):
                self._moveaxes(self.sg.axes[i,j], self.subgrid[i,j])

    def _movejointgrid(self):
        """ Move Jointgrid """
        h= self.sg.ax_joint.get_position().height
        h2= self.sg.ax_marg_x.get_position().height
        r = int(np.round(h/h2))
        self._resize()
        self.subgrid = gridspec.GridSpecFromSubplotSpec(r+1,r+1, subplot_spec=self.subplot)

        self._moveaxes(self.sg.ax_joint, self.subgrid[1:, :-1])
        self._moveaxes(self.sg.ax_marg_x, self.subgrid[0, :-1])
        self._moveaxes(self.sg.ax_marg_y, self.subgrid[1:, -1])

    def _moveaxes(self, ax, gs):
        ax.remove()
        ax.figure=self.fig
        self.fig.axes.append(ax)
        self.fig.add_axes(ax)
        ax._subplotspec = gs
        ax.set_position(gs.get_position(self.fig))
        ax.set_subplotspec(gs)

    def _finalize(self):
        plt.close(self.sg.fig)
        self.fig.canvas.mpl_connect("resize_event", self._resize)
        self.fig.canvas.draw()

    def _resize(self, evt=None):
        self.sg.fig.set_size_inches(self.fig.get_size_inches())
```

```python
# функция для построения графиков распределения тестовой и обучающей выборки
def draw_jointplots(true_y, pred_y):

    sns.set(style='white')

    g0 = sns.jointplot( 
        x=np.array(true_y['Depth']), 
        y=pred_y[:, 0], 
        space=0,
        kind='reg', 
        )
    g0.set_axis_labels('Depth', '', fontsize=16)

    g1 = sns.jointplot( 
        x=np.array(true_y['Width']), 
        y=pred_y[:, 1], 
        space=0,
        kind='reg', 
        )
    g1.set_axis_labels('Width', '', fontsize=16)

    fig = plt.figure(figsize=(12,7))
    fig.suptitle('Распределение прогнозных и фактических значений целевых переменных')
    gs = gridspec.GridSpec(1, 2)

    mg0 = SeabornFig2Grid(g0, fig, gs[0])
    mg1 = SeabornFig2Grid(g1, fig, gs[1])

    gs.tight_layout(fig)

    plt.show()
```

```python
# будем сохранять оценку для моделей
results = []
results = pd.DataFrame(columns=['R2', 'MAE', 'MSE'])

# функция для оценки качества построенной модели
def evaluate(model, test_x, true_y):
    
    model_name = type(model).__name__    
    if model_name == 'MultiOutputRegressor':
        model_name = type(model.estimators_[0]).__name__
        
    pred_y = model.predict(test_x)
    
    mae = mean_absolute_error(true_y, pred_y, multioutput='raw_values')
    mse = mean_squared_error(true_y, pred_y, multioutput='raw_values')
    r2 = r2_score(true_y, pred_y, multioutput='raw_values')
    
    a_mae = mean_absolute_error(true_y, pred_y, multioutput='uniform_average')
    a_mse = mean_squared_error(true_y, pred_y, multioutput='uniform_average')
    a_r2 = r2_score(true_y, pred_y, multioutput='uniform_average')
    
    print(f'Оценка качества модели {model_name}')
    print('―' * 50)
    print('\nR2 / Коэффициент детерминации:')
    print('- для двух целевых переменных', r2)
    print(f'- усредненный {a_r2:.3f}')      
    print('\nMAE / Средняя абсолютная ошибка:')
    print('- для двух целевых переменных', mae)
    print(f'- усредненный {a_mae:.3f}')
    print('\nMSE / Средняя квадратическая ошибка:')
    print('- для двух целевых переменных', mse)
    print(f'- усредненный {a_mse:.3f}')    

    results.loc[model_name] = [a_r2, a_mae, a_mse]
    draw_jointplots(true_y, pred_y)
```

```python
# функция для отрисовки дерева в текстовом и графическом виде
def draw_tree(model, n_est, feature_names):
    text_tree = export_text(model.estimators_[n_est], feature_names=feature_names)
    print('―' * 55)
    print(f'Визуальное отображение {n_est}-го из деревьев случайного леса')
    print('―' * 55)
    print(text_tree)
    dot_data = StringIO()
    export_graphviz(model.estimators_[n_est], out_file='tree.dot', feature_names = feature_names,
                rounded = True, proportion = False, precision = 2, filled = True)
    call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])
    display(Image(filename = 'tree.png'))
```

```python
# функция для вывода значимости признаков в модели
def show_importances(model):
    importances = pd.DataFrame({'feature':features,'importance':np.round(model.feature_importances_,3)})
    importances = importances.sort_values('importance',ascending=False).set_index('feature')
    return importances.T

# функция для вывода весов признаков в модели
def show_weights(model):
    weights = pd.DataFrame(model.coef_, labels, columns=features)
    return weights
```

```python
def plot_loss(history):
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.xlabel('Epoch')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(True)
```

### RandomForestRegressor

Модель "дерево решений" примечательна тем, что годится как для классификации, так и для регрессии — в этом случае в листьях дерева будут числа, а не значение класса. Деревья не требуют серьёзной предобработки данных: они не чувствительны к масштабу признаков и устойчивы к их мультиколлинеарности.
С другой стороны, деревья склонны переобучаться — слишком сильно подгонять параметры под обучающую выборку, и плохо работать на новых данных. Поэтому, чтобы получить более сильную модель, мы воспользуемся алгоритмом случайного леса **RandomForestRegressor**. Он генерирует множество различных независимых друг от друга деревьев слегка разными способами (берёт разные подвыборки, разные признаки), усредняет ответы всех деревьев и формирует итоговое решение. Такой подход может помочь компенсировать ошибки и случайности отдельных деревьев, и получить более точный прогноз.

Построим модель:

```python
# задаём алгоритм для модели 
RFR_model = RandomForestRegressor(random_state=0)
# обучим модель
RFR_model.fit(X_train, Y_train) 
print('\nМодель построена\n')
# сделаем прогноз обученной моделью, получим оценку его точности и нарисуем графики
evaluate(RFR_model, X_test, Y_test)
draw_tree(RFR_model, 0, features)
```

Модель случайного леса позволяет оценить относительную важность каждого признака, таким образом что их сумма равна 1. Выведем эту характеристику для построенной модели:

```python
show_importances(RFR_model)
```

Как видим, наиболее значимым оказался признак `VW` (скорость сварки). Его важность оценивается как почти в 7 раз б**о**льшая, чем следующего признака `IF` (ток фокусировки электронного пучка). Остальные признаки оказались *сравнительно* не важны для построения модели.

### Модели линейной регрессии

При работе с линейной регрессией мы должны учитывать два момента:
- признаки следует привести к единому масштабу, используя нормализацию или стандартизацию, чтобы их веса можно было сравнивать;
- модель линейной регрессии склонна к переобучению и неустойчива в случае взаимной корреляции признаков — на выходе можно получить вектор весов, среди которых будут аномально высокие по модулю и с разными знаками.

Для решения первой проблемы мы ранее провели стандартизацию методом **StandardScaler()** — и для обучения модели в дальнейшем будем использовать стандартизированные данные. 

Что касается второго обстоятельства — мультиколлинеарности входных признаков, есть несколько подходов, которые мы можем применить:
1) оставить только признаки, корреляция между которыми не превышает высокого порога (скажем, 0.8);
2) использовать регуляризацию, то есть ограничение (по сумме модулей или по сумме квадратов) весов в модели.

Второй вариант кажется более предпочтительным, тем более что в библиотеке **sklearn** есть алгоритмы линейной регрессии, которые уже включают в себя регуляризацию:
- **Lasso regression** использует L1-регуляризацию. Этот алгоритм «зануляет» веса при всех сильно скоррелированных признаках, кроме одного. Таким образом, в алгоритм встроен механизм отбора признаков (feature selection) — его применяют, когда нужно снизить размерность и избавиться от дублирующих признаков.
- **Ridge regression** использует L2-регуляризацию. В этом случае веса между скоррелированными признаками будут распределены примерно равномерно.

Попробуем построить обе модели и посмотрим, есть ли разница.

#### Lasso regression

```python
# задаём алгоритм для модели 
Lasso_model = Lasso(alpha=0.002, random_state=0)
# обучим модель
Lasso_model.fit(X_train_st, Y_train) 
print('\nМодель построена\n')
# сделаем прогноз обученной моделью, получим оценку его точности и нарисуем графики
evaluate(Lasso_model, X_test_st, Y_test)
```

```python
show_weights(Lasso_model)
```

Выведем веса, которые модель назначила исходным признакам:

### GradientBoostingRegressor

Градиентный бустинг концептуально похож на лес деревьев, с той разницей, что в этом случае компенсация ошибок происходит не за счёт усреднения, а через обучение. В градиентном бустинге деревья обучаются последовательно: каждое следующее дерево строится с учётом результатов предыдущего. 

Посмотрим, даст ли этот алгоритм лучший результат.

```python
# задаём алгоритм для модели 
GBR_model = MultiOutputRegressor(GradientBoostingRegressor(random_state=0))
# обучим модель
GBR_model.fit(X_train, Y_train) 
print('\nМодель построена\n')
# сделаем прогноз обученной моделью, получим оценку его точности и нарисуем графики
evaluate(GBR_model, X_test, Y_test)
```

Модель градиентного бустинга, как и случайного леса, позволяет оценить относительную важность каждого признака. Выведем ее для построенной модели:

```python
for num, est in enumerate(GBR_model.estimators_):
    print('\nВажность признаков для целевой переменной', labels[num])
    display(show_importances(est))
```

### SVR

Метод опорных векторов библиотеки **sklearn** отличается от использованных выше тем, что нативно не поддерживает мульти-таргет регрессию. Поэтому для прогнозирования двух целевых признаков одновременно будем использовать **MultiOutputRegressor()**.

```python
# задаём алгоритм для модели 
SVR_model = MultiOutputRegressor(SVR(epsilon=0.03))
# обучим модель
SVR_model.fit(X_train_st, Y_train) 
print('\nМодель построена\n')
# сделаем прогноз обученной моделью, получим оценку его точности и нарисуем графики
evaluate(SVR_model, X_test_st, Y_test)
```

Модель показала хороший результат, но не превзошла ранее построенные модели.

### KNeighborsRegressor

Принцип, лежащий в основе метода k-ближайших соседей, состоит в том, чтобы найти определенное количество обучающих выборок, ближайших по расстоянию к новой точке, и предсказать метку — или, в случае задачи регрессии, значение целевого признака — по ним. Прибегнем к этому методу для решения нашей задачи.

```python
# задаём алгоритм для модели 
KNR_model = KNeighborsRegressor(weights='distance')
# обучим модель
KNR_model.fit(X_train_st, Y_train) 
print('\nМодель построена\n')
# сделаем прогноз обученной моделью, получим оценку его точности и нарисуем графики
evaluate(KNR_model, X_test_st, Y_test)
```

Модель **KNeighborsRegressor** показала результат на уровне двух других топовых моделей — **RandomForestRegressor** и **GradientBoostingRegressor**.

### Нейронная сеть

Построим простую полносвязную нейронную сеть. Для решения задачи регрессии зададим выходной слой с двумя линейными нейронами, входной слой и два скрытых слоя с функцией активации *relu*. 

```python
NN_model = Sequential()

# входной слой
NN_model.add(Dense(256, kernel_initializer='normal', input_dim = X_train.shape[1], activation='relu'))

# скрытые слои
NN_model.add(Dense(128, kernel_initializer='normal', activation='relu'))
NN_model.add(Dense(64, kernel_initializer='normal', activation='relu'))

# выходной слой
NN_model.add(Dense(2, kernel_initializer='normal', activation='linear'))

# создадим сеть
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])

# выведем ее архитектуру
NN_model.summary()
```
